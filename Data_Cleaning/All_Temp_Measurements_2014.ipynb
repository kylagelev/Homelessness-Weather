```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
finalAllCityTempsDf=pd.DataFrame([])
cityList=['Bernalillo_County','Clark_County','Cook_County','Dallas_County','Davidson_County','Denver_County','Fulton_County','Hennepin_County','Jackson_County','King_County','Los_Angeles_County','Maricopa_County','Marion_County','Mecklenburg_County','Miami_Dade_County','New_York_County','Philadelphia_County','San_Francisco_County','Suffolk_County','Wayne_County']
```

```python
for city in cityList:    
    #Import CSV file
    csvCity=f"{city}.csv"
    cityDf=pd.read_csv(csvCity)

    #Selected columns we want in our dataframe

    newColumns=['DATE', 'HourlyDryBulbTemperature']

    #Create new data frame based on given columns list

    shortCityDf=cityDf[newColumns]

    #Drop rows with null values. We do not fill values with zeros here as they might be mistaken for temperatures in further analysis

    shortCityDf=shortCityDf.dropna(axis=0)

    #Get only those rows from the month of January, create new data frame for just January

    CityDf2014=shortCityDf.loc[shortCityDf['DATE'].str.contains('2014-01'), newColumns]

    #Convert temperatures into floats. Non-numerical values will be converted to NaN using 'coerce'
    CityDf2014['Hourly_Dry_Bulb_Temperature']=pd.to_numeric(CityDf2014['HourlyDryBulbTemperature'], errors='coerce')

    #Drop the na values

    CityDf2014=CityDf2014.dropna(axis=0)

    countyName=csvCity.replace('.csv', '')

    CityDf2014['County_Name']=countyName

    CityDf2014=CityDf2014.rename(columns={'DATE':'Date'})

    CityDf2014=CityDf2014[['County_Name', 'Date', 'Hourly_Dry_Bulb_Temperature']]

    finalAllCityTempsDf=finalAllCityTempsDf.append(CityDf2014)

finalAllCityTempsDf
```

```python
finalAllCityTempsDf.to_csv('2014_All_Temps.csv', index=False, header=True)
```
