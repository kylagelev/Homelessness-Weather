```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
#import cleaned homelessnes, population and weather data
populationCsv='../county_population.csv'
homelessnessCsv='../Homeless_2015.csv'
weatherCsv='2015_Weather_Summary.csv'
populationFullDf=pd.read_csv(populationCsv)
homelessnessDf=pd.read_csv(homelessnessCsv)
weatherDf=pd.read_csv(weatherCsv)
```

```python
#set index to county
weatherDf=weatherDf.set_index('County')
weatherDf
```

```python
#set homeless data to county and rename columns
homelessnessDf=homelessnessDf.set_index('County')
homelessnessDf=homelessnessDf.rename(columns={'Value_2015':'Total_Homeless'})
homelessnessDf
```

```python
#Merge homelessness and weather data on county name
weatherHomelessDf=pd.merge(weatherDf, homelessnessDf, on='County', how='outer')
weatherHomelessDf
```

```python
#Drop unneeded columns from our population data frame
populationDf=populationFullDf.drop(columns=['REGION', 'DIVISION', 'STATE', 'COUNTY', 'CENSUS2010POP', 'POPESTIMATE2015'])
populationDf
```

```python
#Replace county names in population dataframe with same formatting used in weather and homelessness data frame
populationDf['CTYNAME']=populationDf['CTYNAME'].replace({"Bernalillo County":"Bernalillo_County", "Clark County":"Clark_County", "Cook County":"Cook_County", "Dallas County":"Dallas_County", "Davidson County":"Davidson_County", "Denver County":"Denver_County", "Fulton County":"Fulton_County", "Hennepin County":"Hennepin_County", "Jackson County":"Jackson_County", "King County":"King_County", "Los Angeles County":"Los_Angeles_County", "Maricopa County":"Maricopa_County", "Marion County":"Marion_County", "Mecklenburg County":"Mecklenburg_County", "Miami-Dade County":"Miami_Dade_County", "New York County":"New_York_County", "Philadelphia County":"Philadelphia_County", "San Francisco County":"San_Francisco_County", "Suffolk County":"Suffolk_County", "Wayne County":"Wayne_County"})
populationDf 
```

```python
#Change column headers
populationDf=populationDf.rename(columns={'CTYNAME':'County', 'POPESTIMATE2015':'2015_Population_Estimate'})
populationDf
```

```python
#Set population dataframe index to county
populationDf=populationDf.set_index('County')
populationDf
```

```python
#merge population data with weather and homelessness data
weatherHomelessPopulationDf=pd.merge(weatherHomelessDf, populationDf, on='County', how='outer')
weatherHomelessPopulationDf
```

```python
totalHomeless=weatherHomelessPopulationDf['Total_Homeless'].tolist()
populationEstimate=weatherHomelessPopulationDf['2014_Population_Estimate'].tolist()
homelessRate = [100*(h / p) for h, p in zip(totalHomeless, populationEstimate)]
weatherHomelessPopulationDf['Homelessness_Rate(%)']=homelessRate
```

```python
#Rearrage columns in a more intuitive order
weatherHomelessPopulationDf=weatherHomelessPopulationDf[['Year', 'Homelessness_Rate(%)', 'Total_Homeless', '2015_Population_Estimate', 'Jan_Median_Temp', 'Jan_Mean_Temp', 'Jan_Min_Temp', 'Jan_Max_Temp']]
weatherHomelessPopulationDf
```

```python
#Export to csv
weatherHomelessPopulationDf.to_csv('../2015_Master_Summary.csv', index=True, header=True)
```
