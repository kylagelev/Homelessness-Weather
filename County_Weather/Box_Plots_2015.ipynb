```python
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
csv2014='2014_All_Temps.csv'
csv2015='2015_All_Temps.csv'
```

```python
weather2014Df=pd.read_csv(csv2014)
weather2015Df=pd.read_csv(csv2015)
temps2014Df=weather2014Df[['County_Name', 'Hourly_Dry_Bulb_Temperature']]
temps2015Df=weather2015Df[['County_Name', 'Hourly_Dry_Bulb_Temperature']]
temps2014Df.head()
```

```python
countyList=['Bernalillo','Clark','Cook','Dallas','Davidson','Denver','Fulton','Hennepin','Jackson','King','Los Angeles','Maricopa','Marion','Mecklenburg','Miami Dade','New York','Philadelphia','San Francisco','Suffolk','Wayne']
tickLocations=[value for value in countyList]
```

```python
boxplot=temps2015Df.boxplot(column='Hourly_Dry_Bulb_Temperature',by='County_Name', rot=90, grid=False, figsize=(10, 10), color=dict(boxes='c', whiskers='c', medians='r', caps='c'))
plt.title('Hourly Temperature By County 2015', fontsize=20)
plt.xlabel('County', fontsize=15)
plt.ylabel('Temperature in Deg F', fontsize=15)
plt.suptitle('')
boxplot
```
