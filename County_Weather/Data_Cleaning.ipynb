```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
csvClark="Clark_County.csv"
clarkDf=pd.read_csv(csvClark)
```

```python
columnsList=clarkDf.columns.tolist()
```

```python
columnsList
```

```python
newColumns=['DATE', 'HourlyDryBulbTemperature','MonthlyMaximumTemperature', 'MonthlyMeanTemperature']
shortClarkDf=clarkDf[newColumns]
shortClarkDf
```

```python
ClarkDf2014=shortClarkDf.loc[shortClarkDf['DATE'].str.contains('2014-01'), newColumns]
ClarkDf2014=ClarkDf2014.fillna(value=0)
ClarkDf2014
```

```python
ClarkDf2014MonthlyMedian=ClarkDf2014['HourlyDryBulbTemperature'].median()
ClarkDf2014MonthlyMedian
```

```python
ClarkDf2014Summary=ClarkDf2014.loc[ClarkDf2014['MonthlyMaximumTemperature']!=0,:]
ClarkDf2014Summary
```

```python
ClarkDf2014Summary['MonthlyMedianTemperature']=ClarkDf2014['HourlyDryBulbTemperature'].median()
ClarkDf2014Summary
```

```python
ClarkDf2014Summary=ClarkDf2014Summary.drop('HourlyDryBulbTemperature', axis=1)
ClarkDf2014Summary
```

```python
countyName=csvClark.replace('.csv', '')
ClarkDf2014Summary['County']=countyName
ClarkDf2014Summary.set_index('County')
ClarkDf2014Summary
```
