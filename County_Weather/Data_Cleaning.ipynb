```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
csvClark="Clark_County.csv"
clarkDf=pd.read_csv(csvClark)
```

```python
columnsList=clarkDf.columns.tolist()
```

```python
columnsList
```

```python
newColumns=['STATION', 'DATE', 'HourlyDryBulbTemperature','DailyAverageDryBulbTemperature', 'DailyMaximumDryBulbTemperature','MonthlyMaximumTemperature', 'MonthlyMeanTemperature']
shortClarkDf=clarkDf[newColumns]
shortClarkDf
```

```python
ClarkDf2014=shortClarkDf.loc[shortClarkDf['DATE'].str.contains('2014-01'), newColumns]
ClarkDf2014=ClarkDf2014.fillna(value=0)
ClarkDf2014
```

```python
ClarkDf2014DailyAverage=ClarkDf2014.loc[ClarkDf2014['DailyAverageDryBulbTemperature']!=0,:]
ClarkDf2014DailyAverage
```

```python
ClarkDf2014MonthlyAverage=ClarkDf2014.loc[ClarkDf2014['MonthlyMeanTemperature']!=0, :]
ClarkDf2014MonthlyAverage
```

```python
ClarkDf2014MonthlyMedian=ClarkDf2014['HourlyDryBulbTemperature'].median()
ClarkDf2014MonthlyMedian
```
